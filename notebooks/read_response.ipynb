```python
import pandas as pd
import pickle
```

```python
# Save to dataframe
# Columns: zip code | BR # | avg rent | min | max | total rentals | average per person
row_list = []

for zip in [11206, 11205, 11211, 11222]:
    with open(f'../data/by_zip/rent_agg_by_zip_{zip}.pkl', 'rb') as f:
        response = pickle.load(f)

    for item in response.json()['rentalData']['detailed']:
        if item['bedrooms'] != 0:
            row_dict = {'zip': zip}

            row_dict['bedrooms'] = item['bedrooms']
            row_dict['average_rent'] = item['averageRent']
            row_dict['min_rent'] = item['minRent']
            row_dict['max_rent'] = item['maxRent']
            row_dict['total_rentals'] = item['totalRentals']

            row_dict['average_rent_per_person'] = item['averageRent'] / item['bedrooms']

            row_list.append(row_dict)

df_rents = pd.DataFrame(row_list)
```
